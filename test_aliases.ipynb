# Alias test notebook

This notebook loads aliases from `gutenberg_aliases.txt` (created by the main notebook) or falls back to `gutenberg_pkg.list_aliases()` if available. It prints counts and a small sample and will assert that at least one alias exists.

```python
from pathlib import Path
import sys

# prefer reading the aliases file if it exists
alias_file = Path('gutenberg_aliases.txt')
aliases = []
if alias_file.exists():
    aliases = [l.strip() for l in alias_file.read_text(encoding='utf-8').splitlines() if l.strip()]
else:
    # fall back to package function if present - insert src to path for local dev layout
    sys.path.insert(0, 'src')
    try:
        from gutenberg_pkg import list_aliases
        aliases = list_aliases()
    except Exception as e:
        print('Could not load aliases from package:', e)

print('Total aliases found:', len(aliases))
print('Sample (first 10):', aliases[:10])
assert len(aliases) > 0, 'No aliases found; run the alias-extraction cell in gutenberg.ipynb first.'
```
